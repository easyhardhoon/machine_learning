**2. (x,)에 대해**

기본적으로, (,1) 이런것은 없다. (1,)라는 형식이 있는데,

이는 예를 들어 a = np.array([1,2,3])이면

a.shape를 하면 a의 크기를 **튜플** 형태로 반환하기 떄문이다.

그래서 a.shape는 (3,)이다. **(위의 사진 참고)**

그렇다면 (3,) 과 (1,3) 과  (3,1)은 다른걸까?

코드로 구현해본다



```python
import numpy as np
a = np.array([[1],[2],[3]])
b = np.array([1,2,3])
c = np.array([1])
print(a.shape)
print(b.shape)
print(c.shape)
```

즉,

(3,1)은 y축 3개, x축 1개의 데이터이고

(3,)은 (1,3)과 같은 말이며 y축 1개, x축 3개 데이터라고 할 수 있다.

(3,1)은 2차원 데이터

(3,) == (1,3)은 1차원 데이터이다

형상이 (1,3)인 데이터를 (3,)으로 쓰는 데에는 아래의 이유가 있다

(물론 튜플 형태로 반환하기 때문이긴 하지만)

**3. 행렬곱(np.dot)**



```python
A = np.array([[1,2],[3,4],[5,6]]) #(3,2)
B = np.array([7,8]) #(2,)
np.dot(A,B) # (3,2) dot (2,) = (3,)
```

위의 코드에서 알 수 있듯,

다차원 배열을 곱하려면 두 행렬의 대응하는 차원의 원소 수를 일치시켜야한다

---

⚾(3,2) dot (2,4) --> (3,4)



⚾(3,2) dot (2,) --> (3,)

⚾(2,) dot (2,3) --> (3,)

**4.출력층의 활성화 함수**

일반적으로 

**회귀**의 문제에서는 출력층의 활성화 함수에 **항등함수**를 쓰고

**2클래스 분류**에는 **시그모이드 함수**를

**다중 클래스 분류**에는 **소프트맥스 함수**를 쓴다

---

⚡소프트맥스 함수는 단조 증가 함수이기 떄문에,

소프트맥스 함수를 이용해도 대소 관계는 변하지 않음

그래서 신경망으로 분류할때는 출력층의 소프트맥스 생략 가능.

(학습때에만 사용, 추론[분류]때에 미사용)

---

```python
def softmax(a):
  c = np.max(a)
  exp_a = np.exp(a-c) 
  sum_exp_a = np.sum(exp_a)
  y = exp_a/sum_exp_a
  return y
#소프트맥스의 인자에 어떤 정수를 더해도 결과는 바뀌지 않음을 이용, (c = np.max(a))
#지수함수로 인한 오버플로 방지

def sigmoid(x):
  return 1/(1+np.exp(-x))
  
def identity_function(x):
  return x
```
